```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad,trapezoid,simpson
```

```python
# DELUPPGIFT 1
def v(t,g,w,w0):
    return np.abs(((g * w0**2)/(w0**2  - w**2) * (w0 * np.sin(w0 * t) - w * np.sin(w * t))))
```

```python
w = 2 * np.pi
w0 = 3 * np.pi
g = 0.1 
t = 0
T = 10
N = 4000
time = np.linspace(t,T, N)
```

```python
plt.plot(time, v(time,g,w,w0), label="Hastighet")
plt.title("Hastighet ver tid fr T = 10")
plt.xlabel("Tid[s]")
plt.ylabel("Hastighet [m/s]")
plt.grid()
plt.show()
```

```python
I,err = quad(v,t,T, args = (g,w,w0), limit=500,epsabs=1.49e-12, epsrel=1.49e-12,)
print("Integral fr T = 10:", I)
```

```python
T = 0.25
N = 200
time = np.linspace(t, T, N)
plt.plot(time, v(time, g, w, w0))
plt.title("Hastighet ver tid fr T = 0.25")
plt.xlabel("Tid [s]")
plt.ylabel("Hastighet [m/s]")
plt.grid()
plt.show()

I, err = quad(v,t,T, args = (g,w,w0), limit=500,epsabs=1.49e-12, epsrel=1.49e-12,)
print("Integral fr T = 0.25:", I)
```

```python
# DELUPPGIFT 2
# Trapazoid
T = 0.25
N = 100
time = np.linspace(t,T, N)
Integrals = []
for _ in range(4):
    Integrals.append((float(trapezoid(v(time,g,w,w0), time)), float(trapezoid(v(np.linspace(t,T,int(0.5 * N)), g,w,w0), np.linspace(t,T,int(0.5 * N))))))
    N *= 2
    time = np.linspace(t,T,N)
print(Integrals)
```

```python
# Simpson
T = 0.25
N = 100
time = np.linspace(t,T, N)
Integrals = []
for _ in range(3):
    Integrals.append((float(simpson(v(time,g,w,w0), time)), float(simpson(v(np.linspace(t,T,int(0.5 * N)), g,w,w0), np.linspace(t,T,int(0.5 * N))))))
    N *= 2
    time = np.linspace(t,T,N)
print(Integrals)
```
