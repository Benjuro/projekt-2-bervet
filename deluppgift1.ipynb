```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad,trapezoid 
```

```python
# DELUPPGIFT 1
def v(t,g,w,w0):
    return np.abs(((g * w0**2)/(w0**2  - w**2) * (w0 * np.sin(w0 * t) - w * np.sin(w * t))))
```

```python
w = 2 * np.pi
w0 = 3 * np.pi
g = 0.1
t = 0
T = 10
N = 4000
time = np.linspace(t,T, N)
```

```python
plt.plot(time, v(time,g,w,w0), label="tst")
plt.title("")

plt.legend()
plt.grid()
plt.show()
```

```python
# DELUPPGIFT 2
I,err = quad(v,t,T, args = (g,w,w0), limit=500,epsabs=1.49e-12, epsrel=1.49e-12,)
print(I)
```
